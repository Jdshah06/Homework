```python
import pandas as pd
```

```python
file="Resources/purchase_data.csv"
```

```python
hero_df= pd.read_csv(file)
hero_df
```

```python
#Player Count
players_df= hero_df['SN'].nunique()
players= pd.DataFrame({'Total Players':[players_df]})
players
# players_df.count()
# total_players_1= players_df.value_counts()
# total_players
```

```python
#Unique Items
items_df= hero_df['Item ID'].unique()
# item_count=items_df.nunique()
unique_items=len(items_df)
#Number of Purchases
p_total= hero_df['Purchase ID']
n_purchases= len(p_total)
#Average Price 
price_df= hero_df['Price']
avg_price=price_df.mean()
avg_price
# Total Purchase Value
purchase_df= hero_df['Price']
total_purchases= purchase_df.sum()
total_purchases


Purchase_Analysis=pd.DataFrame({"Unique Items":[unique_items],
                                 "Average Price":[avg_price],
                                "Number of Purchases":[n_purchases],
                                 "Total Purchases":[total_purchases]})
Purchase_Analysis
```

```python
gender=['Female','Male','Other/Non-disclosed']
```

```python
#Female Purchasing Analysis
female_df= hero_df.loc[hero_df['Gender'] == "Female", :]
female_count=female_df['SN'].nunique()
# female_df
female_purchase_count= female_df['Purchase ID'].count()
# female_purchase_count
female_avg_price= female_df['Price'].mean()
#female_average_price
female_purchase_value= female_df['Price'].sum()
#female_total_purchase_value
female_avg_per_person= female_purchase_value/female_count
#female_avg_per_person
# print(female_purchase_count)
# print(female_avg_price)
# print(female_purchase_value)
# print(female_avg_per_person)
```

```python
# Male Purchasing Analysis
male_df= hero_df.loc[hero_df['Gender'] == "Male", :]
male_count=male_df['SN'].nunique()
male_purchase_count= male_df['Purchase ID'].count()
male_avg_price= male_df['Price'].mean()
male_purchase_value= male_df['Price'].sum()
male_avg_per_person= male_purchase_value/male_count
# print(male_purchase_count)
# print(male_avg_price)
# print(male_purchase_value)
# print(male_avg_per_person)
```

```python
#N/A Purchasing Analysis
other_df= hero_df.loc[hero_df['Gender'] == "Other / Non-Disclosed", :]
other_purchase_count= other_df['Purchase ID'].count()
other_avg_price= other_df['Price'].mean()
other_purchase_value= other_df['Price'].sum()
other_avg_per_person= other_purchase_value/other_avg_price
# print(other_purchase_count)
# print(other_avg_price)
# print(other_purchase_value)
# print(other_avg_per_person)
```

```python
Gender_analysis=pd.DataFrame({"Gender":gender,"Purchase Count":[female_purchase_count,male_purchase_count,other_purchase_count],
                              "Average Purchase Price":[female_avg_price,male_avg_price,other_avg_price],
                              "Total Purchase Value":[female_purchase_value,male_purchase_value,other_purchase_value],
                              "Avg Total Purchase per Person":[female_avg_per_person,male_avg_per_person,other_avg_per_person]})
# Gender_analysis.style.format({"Average Purchase Price":"${:.2f}",
#                                              "Total Purchase Value":"${:.2f}",
#                                              "Avg Total Purchase per Person":"${.2f}"})
Gender_analysis
```

```python
#Age Groups
age_bin= hero_df
youngest=hero_df['Age'].min()
oldest= hero_df['Age'].max()
# print(youngest)
# print(oldest)
bins=[0,10,14,19,24,29,34,39,100]
labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
age_bin['Age Groups']= pd.cut(hero_df['Age'],bins,labels=labels, include_lowest=True)
age_bin=age_bin.groupby(["Age Groups"])
bin_unique= age_bin["SN"].nunique()
Age_analysis= pd.DataFrame({"Purchase Count":age_bin['Purchase ID'].count(),
                            "Average Purchase Price":age_bin['Price'].mean(),
                           "Total Purchase Value":age_bin['Price'].sum(),
                           "Avg Total Purchase Per Age Group":
                            age_bin['Price'].sum()/bin_unique})
Age_analysis=Age_analysis.style.format({"Average Purchase Price":"${:.2f}",
                                             "Total Purchase Value":"${:.2f}",
                                             "Avg Total Purchase Per Age Group":"${:.2f}"})
Age_analysis
```
